# Search / Query

```python
nums = [1, 2, 3, 4, 5]
nums_series = pd.Series(nums, name="Numbers")
nums_series[nums_series >= 3]
```

```python
animals = ["Cat", "Dog", "Pig", "Cow", None]
animal_series = pd.Series(animals, name="Animals")
animal_series[animal_series=="Cat"]
```

```python
animal_series[animal_series.fillna('').str.startswith("C")]
```
