# Loop

## For-Loop

```python
numbers = [1, 2, 3, 4, 5]
for n in numbers:
    print("===============")
    print(n)
print("Finished")
```

```python
tuple_numbers = (1, 2, 3, 4, 5)
for n in tuple_numbers:
    print("===============")
    print(n)
print("Finished")

```

```python
person = { 'name': "Sambath", "age": 20}
for value in person.keys():
    print(value)
```

```python
person = { 'name': "Sambath", "age": 20}
for value in person.values():
    print(value)
```

```python
for v, k in person.items():
    print(v, k)
```

```python
tuple_numbers = (1, 2, 3, 4, 5)
accumulated_total = 0
for ele in tuple_numbers:
    accumulated_total = accumulated_total + ele
accumulated_total
```

```python
tuple_numbers = (1, 2, 3, 4, 5)
total = 0
for ele in tuple_numbers:
    print(ele)
total = total + ele
```

## While loop

```python
for i in range(100):
    print(i)
```

```python
loop = 100
step = 1
while step <= loop:
    print(step)
    step = step + 1
```

```python
age = input("Enter your age: ")
continue_work = True
while continue_work:
    try: 
        age = int(age)
        continue_work = False
    except:
        print("Invalid input, please enter a valid age")
        age = input("Enter your age: ")
print(f"You age is {age}")
```

# Function

```python
def calculate_area(width, height):
    print("width", width)
    print("height", height)
    print(f"The area is {width * height}")
```

```python
calculate_area()
```

```python
calculate_area(10, 200)
```

```python
area = calculate_area(10, 200)
print("area", area)
```

```python
def calculate_area_with_return(width, height):
    print("width", width)
    print("height", height)
    print(f"The area is {width * height}")
    return width * height

area = calculate_area_with_return(10, 200)
area
```

```python
calculate_area_with_return(height=10, width=200)
```
